```python
#Discuss string slicing and provide example
'''String slicing in Python refers to extracting a part of a string using its index positions. It allows you to access a subset of the characters from a string by specifying a starting and ending position. The syntax for string slicing is:'''

'''string[start:stop:step]'''

'''start: The starting index (inclusive, defaults to 0 if not specified)
stop: The ending index (exclusive, so the character at this index is not included)
step: The step value (optional, defaults to 1, which means characters are taken consecutively)

Here are some examples to demonstrate string slicing:'''

#Example 1: Basic Slicing

s = "Hello, World!"
print(s[0:5])

'''In this case, the slice starts at index 0 and goes up to, but does not include, index 5. So it extracts "Hello".'''

#Example 2: Omitting the start or stop values

print(s[:5])
print(s[7:])

#Omitting the start defaults to the beginning of the string. Omitting the stop defaults to the end of the string.

#Example 3: Using a Negative Index

print(s[-6:])

'''Negative indices count from the end of the string.
Here, -6 starts slicing from the 6th character from the end of the string.'''

#Example 4: Using a step value

print(s[::2])

#The step value of 2 means every second character is taken, starting from the beginning.

#Example 5: Reversing a String

print(s[::-1])

#A step value of -1 reverses the string by starting from the end and moving backward.

#Example 6: Slicing with a Negative start and stop

print(s[-6:-1])

#Here, slicing starts from index -6 (which is the 'W') and ends at -1 (just before the '!').
```

```python
#Explain the key features of lists in Python.
'''Lists in Python are a versatile and commonly used data structure. They are dynamic, mutable, and can store an ordered collection of items, which makes them powerful for handling a variety of use cases. Here are the key features of lists in Python:

 1. Ordered
- Lists maintain the order of elements, meaning the order in which elements are inserted is preserved.
- Elements can be accessed by their index positions.'''


my_list = [1, 2, 3, 4]
print(my_list[0])


 #2. Mutable
'''- Lists are mutable, meaning the elements within a list can be modified after the list is created.
- You can change, add, or remove elements in a list.'''


my_list[2] = 10   # Changing the value at index 2
print(my_list)


'''3. Heterogeneous
- Lists can store elements of different data types (integers, strings, floats, objects, etc.).'''


mixed_list = [1, "apple", 3.14, True]
print(mixed_list)


'''4. Dynamic Size
- Lists in Python are dynamic, which means their size can change during runtime (elements can be added or removed).'''

my_list.append(5)   # Adds 5 to the list
print(my_list)


'''5. Supports Duplicate Elements
- Lists can contain duplicate values; there is no restriction on repeating elements.'''


my_list = [1, 2, 2, 3, 4]
print(my_list)

'''6. Indexing and Slicing
- You can access elements by their index positions, both positive and negative.
- Lists also support slicing, allowing you to extract sublists.'''

my_list = [10, 20, 30, 40, 50]
print(my_list[1:4])  # Output: [20, 30, 40]  (slicing)
print(my_list[-1])

'''7. Built-in List Methods
- Python provides many useful list methods for manipulation, such as:
  - append(): Add an item to the end of the list.
  - extend(): Extend the list by appending all elements from another iterable.
  - insert(): Insert an element at a specific index.
  - remove(): Remove the first occurrence of a specific element.
  - pop(): Remove and return an element from a specific position.
  - sort(): Sort the list in ascending or descending order.
  - reverse(): Reverse the order of the elements in the list.
  - count(): Count occurrences of a value in the list.'''


my_list = [3, 1, 4, 1, 5]
my_list.sort()          # Sorting the list
print(my_list)


'''8. List Comprehensions
- List comprehensions provide a concise way to create lists based on existing lists.
- They are commonly used for generating lists in a more readable and efficient manner.'''

squares = [x**2 for x in range(5)]
print(squares)

'''9. Nesting Lists
- Lists can contain other lists as elements, allowing for the creation of multidimensional lists.'''

matrix = [[1, 2], [3, 4], [5, 6]]
print(matrix[1])       # Output: [3, 4]
print(matrix[1][0])

'''10. Iteration
- Lists support iteration using loops, such as for loops.'''

my_list = [1, 2, 3, 4]
for item in my_list:
    print(item)
```

```python
#Describe the key features of sets and provide examples of their use
'''A *set* in Python is an unordered collection of unique elements. Sets are useful when you need to store items without duplicates and perform mathematical set operations such as unions, intersections, and differences. Here are the key features of sets along with examples of their use:'''

# Key Features of Sets

#1. *Unordered Collection*
'''- Sets are *unordered*, meaning they do not preserve the order in which elements are inserted. The elements can appear in any order when you access them.'''


my_set = {1, 2, 3, 4}
print(my_set)  # Output might be: {1, 2, 3, 4}, but the order is not guaranteed


#2. *Unique Elements*
'''- Sets automatically remove duplicate elements. If you attempt to add a duplicate element to a set, it will only store one copy of that element.'''

my_set = {1, 2, 2, 3, 4}
print(my_set)  # Output: {1, 2, 3, 4} (duplicates are removed)


#3. *Mutable*
'''- Sets themselves are mutable, meaning you can add or remove elements after the set is created. However, the elements within a set must be immutable (e.g., numbers, strings, tuples).'''


my_set = {1, 2, 3}
my_set.add(4)       # Adding an element
print(my_set)       # Output: {1, 2, 3, 4}


#4. *No Indexing or Slicing*
   #- Since sets are unordered, you cannot access elements by index or slice them. You can iterate over a set, but not access elements by position.

my_set = {1, 2, 3}
# print(my_set[0])  # This will raise an error because sets do not support indexing


#5. *Set Operations*
#Python sets support various mathematical set operations like *union, **intersection, **difference, and **symmetric difference*.

'''   - *Union* (| or union()): Combines elements from two sets, removing duplicates.
   - *Intersection* (& or intersection()): Returns elements that are present in both sets.
   - *Difference* (- or difference()): Returns elements in the first set but not in the second.
   - *Symmetric Difference* (^ or symmetric_difference()): Returns elements in either set but not both.'''
set1 = {1, 2, 3}
set2 = {3, 4, 5}

   # Union
print(set1 | set2)  # Output: {1, 2, 3, 4, 5}

   # Intersection
print(set1 & set2)  # Output: {3}

   # Difference
print(set1 - set2)  # Output: {1, 2}

   # Symmetric Difference
print(set1 ^ set2)  # Output: {1, 2, 4, 5}


#6. *Efficient Membership Testing*
'''Checking whether an element is in a set is highly efficient (average time complexity is O(1)), making sets ideal for membership testing.'''

my_set = {1, 2, 3, 4, 5}
print(3 in my_set)  # Output: True
print(6 in my_set)  # Output: False


#7. *Set Comprehensions*
'''Similar to list comprehensions, you can use set comprehensions to create sets dynamically based on certain conditions or transformations.'''


squared_set = {x**2 for x in range(5)}
print(squared_set)  # Output: {0, 1, 4, 9, 16}


#8. *Immutable Sets: frozenset*
'''Python provides an immutable version of a set called a frozenset. Once a frozenset is created, you cannot add or remove elements from it'''


frozen = frozenset([1, 2, 3])
   # frozen.add(4)  # This will raise an AttributeError because frozensets are immutable
print(frozen)    # Output: frozenset({1, 2, 3})


#Examples of Set Use

#Example 1: Removing Duplicates from a List
'''One common use of sets is to remove duplicate elements from a list.'''

my_list = [1, 2, 2, 3, 4, 4, 5]
unique_set = set(my_list)  # Convert list to a set to remove duplicates
print(unique_set)          # Output: {1, 2, 3, 4, 5}


#Example 2: Membership Testing
'''Sets are used when you need to efficiently check if an element exists in a collection.'''

allowed_users = {"alice", "bob", "charlie"}
if "david" in allowed_users:
    print("Access granted")
else:
    print("Access denied")
# Output: Access denied


# Example 3: Mathematical Set Operations
'''You can use sets to perform operations such as finding common or different elements between two collections.'''

set_a = {"apple", "banana", "cherry"}
set_b = {"cherry", "date", "fig"}

# Union
print(set_a | set_b)  # Output: {'banana', 'cherry', 'fig', 'apple', 'date'}

# Intersection
print(set_a & set_b)  # Output: {'cherry'}

# Difference
print(set_a - set_b)  # Output: {'banana', 'apple'}


# Example 4: Filtering Data with Set Comprehensions
'''You can use set comprehensions to filter or transform data.'''

numbers = {1, 2, 3, 4, 5, 6}
even_numbers = {x for x in numbers if x % 2 == 0}
print(even_numbers)  # Output: {2, 4, 6}
```

```python
#Describe how to add, modify, and delete items in a dictionary with examples.
'''Dictionaries in Python are mutable collections that store data in key-value pairs.
You can add, modify, and delete items in a dictionary easily. Below is a detailed description of how to perform these operations along with examples.'''

# 1. *Adding Items to a Dictionary*

'''You can add new key-value pairs to a dictionary by assigning a value to a new key using the assignment (=) operator.'''

# Example:

# Creating an empty dictionary
my_dict = {}

# Adding a key-value pair
my_dict['name'] = 'Jay'
my_dict['age'] = 29

print(my_dict)  # Output: {'name': 'Jay', 'age': 29}


'''If the key already exists, the value will be updated (not duplicated), but if the key is new, it will be added to the dictionary.'''

#2. *Modifying Items in a Dictionary*

'''To modify an existing item in a dictionary, you can simply assign a new value to an existing key.
The old value will be overwritten with the new one.'''

# Example:

# Existing dictionary
my_dict = {'name': 'Jay', 'age': 29}

# Modifying the value associated with the 'age' key
my_dict['age'] = 31

print(my_dict)  # Output: {'name': 'Jay', 'age': 31}


### 3. *Adding/Updating Multiple Items with update()*

'''The update() method can be used to add or update multiple key-value pairs at once.'''

#### Example:

# Existing dictionary
my_dict = {'name': 'Jay', 'age': 29}

# Adding and updating multiple key-value pairs
my_dict.update({'age': 32, 'city': 'Delhi'})

print(my_dict)  # Output: {'name': 'Jay', 'age': 29, 'city': 'Delhi'}


# 4. *Deleting Items from a Dictionary*

'''There are multiple ways to delete items from a dictionary:'''

# a. Using del Statement
'''The del statement removes a key-value pair by specifying the key.'''


my_dict = {'name': 'Jay', 'age': 29, 'city': 'Delhi'}

# Deleting the 'city' key
del my_dict['city']

print(my_dict)  # Output: {'name': 'Jay', 'age': 29}


# b. Using pop() Method
'''The pop() method removes a key-value pair by the key and returns the value that was removed.'''


my_dict = {'name': 'Jay', 'age': 29, 'city': 'Delhi'}

# Deleting the 'age' key using pop()
age = my_dict.pop('age')

print(my_dict)  # Output: {'name': 'Jay', 'city': 'Delhi'}
print(age)      # Output: 29 (the value that was removed)


#c. Using popitem() Method The popitem() method removes and returns the last inserted key-value pair in the dictionary.

'''In Python 3.7 and later, dictionaries maintain insertion order, so this method will remove the most recently added item'''

my_dict = {'name': 'Jay', 'age': 29, 'city': 'Delhi'}

# Deleting the last item using popitem()
last_item = my_dict.popitem()

print(my_dict)  # Output: {'name': 'Jay', 'age': 29}
print(last_item)  # Output: ('city', 'Delhi')


# d. Using clear() Method
#The clear() method removes all items from the dictionary, leaving it empty.

my_dict = {'name': 'Jay', 'age': 29, 'city': 'Delhi'}

# Clearing the entire dictionary
my_dict.clear()

print(my_dict)  # Output: {}


# 5. *Checking and Handling Key Existence*

#Before modifying or deleting a key, it’s a good practice to check if the key exists to avoid errors.

# Example:

my_dict = {'name': 'John', 'age': 31}

# Check if the key 'city' exists before deleting it
if 'city' in my_dict:
  del my_dict['city']
else:
  print("Key 'city' does not exist.")
```

```python
#Discuss the importance of dictionary keys being immutable and provide examples.
'''In Python, dictionary keys must be *immutable* because dictionaries rely on hash values to efficiently store and retrieve key-value pairs.
Mutability would break the consistency of the dictionary and lead to unpredictable behavior.'''

# Why Must Dictionary Keys Be Immutable?

# 1. *Hashing Requirement for Dictionary Keys*
'''- Python dictionaries are implemented as *hash maps*. Each key in a dictionary is hashed using a hash function, which generates an integer (the hash value) that is used to determine where to store the key-value pair in memory.
   - *Immutable objects* such as strings, numbers, and tuples have consistent hash values throughout their lifetime because they don’t change. Therefore, they can serve as reliable dictionary keys.
   - *Mutable objects* like lists or dictionaries can change after being created, which would change their hash value.
   If a key's hash changes after it has been placed in the dictionary, Python would no longer know where to find it, breaking the integrity of the dictionary.'''

# 2. *Consistency and Integrity*
'''- Immutable keys ensure that the mapping between keys and values remains consistent throughout the program. If keys were mutable, the hash could change, making the dictionary unreliable for lookup operations.
   - For example, if a list were used as a key, modifying the list would cause the key to change, and it would no longer be associated with the correct value.'''

# 3. *Efficient Lookups*
'''- Dictionaries are optimized for fast lookups by using hashing. When a key is immutable, the hash value remains constant, allowing Python to quickly compute where the key-value pair is stored in memory.
   - If keys were mutable, this efficiency would be lost because Python would have to compute a new hash each time the key was modified, potentially leading to slow and incorrect lookups.'''

#Immutable Types That Can Be Used as Dictionary Keys

'''*Strings*: Strings are immutable, so they can be used as dictionary keys.'''

my_dict = {'name': 'Sayantani', 'age': 27}
print(my_dict['name'])  # Output: Sayantani


'''*Numbers*: Integers and floats are immutable and are commonly used as keys.'''

my_dict = {1: 'one', 2: 'two'}
print(my_dict[1])  # Output: one


'''*Tuples*: As long as all elements within the tuple are also immutable (e.g., strings, numbers), tuples can serve as dictionary keys.'''

my_dict = {(1, 2): 'coordinates', (3, 4): 'more coordinates'}
print(my_dict[(1, 2)])  # Output: coordinates


### Mutable Types That Cannot Be Used as Dictionary Keys

'''*Lists*: Lists are mutable, meaning their contents can be modified after creation. Attempting to use a list as a key will raise a TypeError.'''


'''
Example:
my_dict = {[1, 2, 3]: 'invalid'}'''  # Raises TypeError: unhashable type: 'list'


'''*Dictionaries*: Like lists, dictionaries are mutable and cannot be used as keys for the same reason.'''


'''my_dict = {{'key': 'value'}: 'invalid'}'''  # Raises TypeError: unhashable type: 'dict'


### Example of Why Mutability Would Cause Issues

'''Consider this scenario where a mutable list is allowed to be a dictionary key:'''

# Hypothetical situation (won't work in Python)
'''my_dict = {[1, 2, 3]: 'value'}'''  # List used as key

# Modify the list
key_list = [1, 2, 3]
key_list.append(4)

# Now the original key's hash has changed and the dictionary will break
'''print(my_dict[key_list])'''  # This would raise an error or return an incorrect result


'''In this hypothetical case, modifying the list would cause its hash value to change.
 The dictionary would not be able to find the correct key-value pair anymore because it depends on the hash value of the original list,
 which no longer exists in its original form.'''

# Example of Proper Immutable Keys

#1. *Using a string as a key* (strings are immutable):

my_dict = {'first_name': 'Sayantani', 'last_name': 'Smith'}
print(my_dict['first_name'])  # Output: Sayantani


#2. *Using a tuple as a key* (tuples are immutable):
my_dict = {(10, 20): 'coordinates', (30, 40): 'more coordinates'}
print(my_dict[(10, 20)])  # Output: coordinates


#3. *Using an integer as a key* (integers are immutable):
my_dict = {1: 'apple', 2: 'banana'}
print(my_dict[1])  # Output: apple
```
