```python
# Describe the role of predefined keywords in python and provide examples of how they are used in a program
"""
Predefined keywords, also known as reserved words, are built-in identifiers that have special meanings in Python. These keywords serve as the core building blocks of the language, defining its syntax and structure. They cannot be used as identifiers (e.g., variable or function names) since they are reserved for their intended purpose.
Key Role of Predefined Keywords:
1.	Control Flow: Keywords such as if, else, elif, while, and for control the execution flow of a Python program based on conditions or loops.
2.	Function Definition and Handling: Keywords like def, return, lambda, and yield are used to define and control functions, including their return values and generator behaviour.
3.	Variable and Object Handling: Keywords like None, True, False, and as handle object states, type checking, and type casting.
4.	Exception Handling: Keywords like try, except, finally, and raise deal with error handling and exceptions in Python programs.
5.	Class and Object-Oriented Programming: Keywords such as class, self, is, in, and with help in defining classes, inheritance, and context management in Python.
6.	Loop and Iteration: Keywords like break, continue, and pass control loop behavior, allowing the program to break out of loops, skip iterations, or serve as a placeholder.
"""
#Examples of Predefined Keywords in Python:
#1. Control Flow:
x = 10
if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")
    
#2. Function Definition and Handling:
def greet(name):
    return f"Hello, {name}!"

print(greet("Sayantani"))
#3. Loop and Iteration:
for i in range(5):
    if i == 3:
        continue
    print(i)
#4. Exception Handling:
try:
    result = 10 / 0
except ZeroDivisionError:
    print("You can't divide by zero!")
finally:
    print("This runs no matter what.")
#•	Keywords: try, except, finally
#•	Role: try defines the block where errors are anticipated, except handles the specific exception, and finally runs code regardless of what happens in the try block.
#5. Class Definition:
class Animal:
    def _init_(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound."

dog = Animal("Dog")
print(dog.speak())
#Reserved Keywords in Python:
"""
•	Control Flow: if, else, elif, while, for, break, continue, pass
•	Functions: def, return, lambda
•	Data Types: None, True, False
•	Error Handling: try, except, finally, raise
"""
```

```python
#Compare and contrast mutable and immutable objects in python with example.
"""
In Python, the terms "mutable" and "immutable" refer to whether an object's state can be changed after it is created. Here's a breakdown of the two concepts with examples:
Mutable Objects
Mutable objects are those whose state or content can be changed after their creation. This means you can modify the object in place without creating a new object.\
"""
#Examples of Mutable Objects:
#Lists:
my_list = [1, 2, 3]
my_list[0] = 10  # Modifying an element in the list
print(my_list)
#Dictionaries:
my_dict = {'key1': 'value1', 'key2': 'value2'}
my_dict['key1'] = 'new_value1'  # Modifying a value in the dictionary
print(my_dict)  
#Sets:
my_set = {1, 2, 3}
my_set.add(4)  # Adding an element to the set
print(my_set) 
#Immutable Objects
"""
Immutable objects, on the other hand, cannot be changed once they are created. Any modification operation will result in the creation of a new object rather than modifying the original one.
"""
#Examples of Immutable Objects:
#Tuples:
my_tuple = (1, 2, 3)
# Attempting to modify an element will raise an error
# my_tuple[0] = 10  # TypeError: 'tuple' object does not support item assignment
#Strings:
my_string = "hello"
new_string = my_string.upper()  # Creating a new string with modifications
print(new_string) 
print(my_string) 
#Numbers:
x = 5
y = x + 1  # Creating a new integer object
print(x)  
print(y)  
#Key Differences
'''
Modification:
1.	Mutable objects can be modified in place.
2.	Immutable objects cannot be modified after creation. Any "modification" results in the creation of a new object.
Memory Usage:
	Mutable objects typically use more memory because they can be altered in place.
	Immutable objects can be optimized by Python's memory management system since they do not change and can be shared safely.
Performance:
1. Mutable objects may have performance overhead associated with their dynamic nature.
2. Immutable objects can be faster in some cases due to their fixed size and state.
'''
```

```python
#Discuss the different type of operators in python and provide examples how they are used.
"""
Python offers several types of operators to perform different operations on variables and values. These operators can be categorized into various types. Below is an overview of the main types of operators in Python with examples:
"""

#1. Arithmetic Operators
#These are used to perform mathematical operations.
#Addition (+): Adds two values.
x = 5 + 3 
#Subtraction (-): Subtracts the second value from the first.
x = 10 - 3
#Multiplication (*): Multiplies two values.
x = 4 * 5  
#Division (/): Divides the first value by the second (result is a float).
x = 9 / 2  
#Floor Division (//): Divides and returns the integer part of the result.
x = 9 // 2
#Modulus (%): Returns the remainder of the division.
x = 9 % 2
#Exponentiation (**): Raises the first number to the power of the second.
x = 2 ** 3
#2. Comparison (Relational) Operators
#These compare two values and return a Boolean result (True or False).
#Equal to (==): Checks if two values are equal.
x = (5 == 5) 
#Not equal to (!=): Checks if two values are not equal.
x = (5 != 3)
#Greater than (>): Checks if the first value is greater than the second.
x = (7 > 3) 
#Less than (<): Checks if the first value is less than the second.
x = (3 < 5) 
#Greater than or equal to (>=): Checks if the first value is greater than or equal to the second.
x = (5 >= 5)
#Less than or equal to (<=): Checks if the first value is less than or equal to the second.
x = (3 <= 5)
#3. Assignment Operators
#These are used to assign values to variables.
#Simple assignment (=):
x = 10  # Assigns 10 to x
#Add and assign (+=):
x = 5
x += 3  # Equivalent to x = x + 3
# x is now 8
#Subtract and assign (-=):
x = 10
x -= 2  # Equivalent to x = x - 2
# x is now 8
#Multiply and assign (*=):
x = 4
x *= 2  # Equivalent to x = x * 2
# x is now 8
#Divide and assign (/=):
x = 9
x /= 3  # Equivalent to x = x / 3
# x is now 3.0
#Modulus and assign (%=):
x = 10
x %= 3  # Equivalent to x = x % 3
# x is now 1
#4. Logical Operators
#These are used to perform logical operations, often in conjunction with comparison operators.
#AND (and): Returns True if both conditions are true.
x = (5 > 3 and 7 > 6)  # True
#OR (or): Returns True if at least one of the conditions is true.
x = (5 > 3 or 7 < 6)  # True
#NOT (not): Reverses the Boolean value of an expression.
x = not(5 > 3)  # False
#5. Bitwise Operators
#These work at the bit level on binary representations of numbers.
#AND (&): Sets each bit to 1 if both bits are 1.
x = 5 & 3  # 1 (binary: 101 & 011 = 001)
#OR (|): Sets each bit to 1 if at least one of the bits is 1.
x = 5 | 3  # 7 (binary: 101 | 011 = 111)
#XOR (^): Sets each bit to 1 if only one of the bits is 1.
x = 5 ^ 3  # 6 (binary: 101 ^ 011 = 110)
#Left Shift (<<): Shifts bits to the left by the specified number of positions.
x = 5 << 1  # 10 (binary: 101 << 1 = 1010)
#Right Shift (>>): Shifts bits to the right by the specified number of positions.
x = 5 >> 1  # 2 (binary: 101 >> 1 = 010)
#6. Identity Operators
#These check whether two variables refer to the same object in memory.
#is: Returns True if both variables refer to the same object.
x = [1, 2, 3]
y = x
z = [1, 2, 3]
print(x is y)  
print(x is z) 
#is not: Returns True if both variables refer to different objects.
print(x is not z)  # True
#7. Membership Operators
#These are used to test if a sequence contains a certain value.
#in: Returns True if a value is found in a sequence.
x = [1, 2, 3]
print(2 in x) 
#not in: Returns True if a value is not found in a sequence.
print(4 not in x)  
#8. Ternary Operator (Conditional Expression)
#This is used to evaluate an expression based on a condition.
x = 10
result = "Positive" if x > 0 else "Negative"
print(result) 
```

```python
#Explain the concept of type casting in python with example.
'''Type casting (also known as type conversion) in Python refers to converting one data type into another. Python provides several built-in functions for type casting to help with operations between incompatible types (like converting strings to numbers or vice versa).
Types of Type Casting
There are two types of type casting in Python:
1.	Implicit Type Casting: Python automatically converts one data type to another.
2.	Explicit Type Casting: You manually convert one data type to another using built-in functions.
#1. Implicit Type Casting
In implicit type casting, Python automatically converts one data type to another without requiring explicit intervention by the programmer. This typically happens when combining types where Python can ensure no loss of data.'''
#Example:
# Implicit type conversion (int to float)
x = 10     # integer
y = 3.5    # float

result = x + y  # Python automatically converts x (int) to float
print(result) 
print(type(result)) 
'''In the example above, Python automatically converts x from an integer to a float to match y (float), and the result is also a float.'''
#2. Explicit Type Casting
'''In explicit type casting, the programmer uses built-in functions to convert one data type into another. This is necessary when you need to convert between incompatible types, such as from a string to an integer, or when Python doesn't perform implicit conversion automatically.'''
'''Common Built-in Type Casting Functions:
int(): Converts a value to an integer.
float(): Converts a value to a float.
str(): Converts a value to a string.
list(): Converts an iterable to a list.
tuple(): Converts an iterable to a tuple.
set(): Converts an iterable to a set.
'''
#Examples:
#1.String to Integer:
#Explicit type conversion from string to int
x = "50"
y = int(x)  # Convert string '50' to integer 50
print(y)
print(type(y))  # Output: <class 'int'>
#In this case, int(x) explicitly converts the string '50' to the integer 50.
#2.Integer to Float:
x = 5
y = float(x)  # Convert integer 5 to float 5.0
print(y)
print(type(y))  # Output: <class 'float'>
#3.Float to Integer:
x = 9.8
y = int(x)  # Convert float 9.8 to integer 9 (decimal part is truncated)
print(y)
print(type(y))  # Output: <class 'int'>
#Note: When converting from float to int, the decimal part is truncated, not rounded.
#4.Integer to String:
x = 100
y = str(x) 
print(y) 
print(type(y))  # Output: <class 'str'>
#5.String to List:
x = "hello"
y = list(x)  # Convert string 'hello' to list of characters ['h', 'e', 'l', 'l', 'o']
print(y) 
print(type(y))
#6.Tuple to List:
x = (1, 2, 3)
y = list(x)  # Convert tuple (1, 2, 3) to list [1, 2, 3]
print(y) 
print(type(y)) 
#7. List to Set:
x = [1, 2, 2, 3, 4]
y = set(x)
print(y)
print(type(y))
#Handling Errors During Type Casting
'''When type casting, it's important to ensure that the conversion is valid. If the conversion is not possible, Python will raise an error.'''
#Example of Error During Type Casting:
x = "abc"
y = int(x)  # Trying to convert string 'abc' to integer will raise a ValueError
#This will raise a ValueError: invalid literal for int() because Python cannot convert the string 'abc' to an integer.
```

```python
#How do conditional statements work in Python? Illustrate with examples.
'''Conditional statements in Python allow the program to execute certain pieces of code based on specific conditions. These conditions are evaluated as True or False, and based on the result, the corresponding block of code is executed.'''
#Types of Conditional Statements in Python:
'''
1.	if statement
2.	if-else statement
3.	if-elif-else statement
4.	Nested if statements
'''
#1. if Statement
#The if statement checks whether a condition is True and executes a block of code if it is.
#Syntax:
#if condition:
    # Block of code to execute if condition is True
#Example:
age = 18
if age >= 18:
    print("You are eligible to vote.")
#In this example, the condition age >= 18 is True, so the message "You are eligible to vote." is printed.
#2. if-else Statement
#The if-else statement provides an alternative block of code to execute when the condition is False.
#Syntax:
#if condition:
    # Block of code to execute if condition is True
#else:
    # Block of code to execute if condition is False
#Example:
age = 16
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote yet.")
#In this case, the condition age >= 18 is False, so the else block is executed.
#3. if-elif-else Statement
#The if-elif-else statement allows multiple conditions to be checked sequentially. Once a condition is True, its corresponding block of code is executed, and the rest of the conditions are skipped.
#Syntax:
#if condition1:
    # Block of code to execute if condition1 is True
#elif condition2:
    # Block of code to execute if condition2 is True
#else:
    # Block of code to execute if all conditions are False
#Example:
marks = 75
if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: F")
'''Here, the second condition marks >= 75 is True, so the corresponding block is executed, and Python skips the rest of the conditions.'''
#4. Nested if Statements
#You can also place if statements inside other if statements to check multiple conditions in a more complex manner.
#Syntax:
#if condition1:
   # if condition2:
        # Block of code to execute if both condition1 and condition2 are True
#Example:
age = 20
citizenship = "US"
if age >= 18:
    if citizenship == "US":
        print("You are eligible to vote in the US.")
    else:
        print("You are not a US citizen.")
else:
    print("You are not old enough to vote.")
#In this case, the program checks if the person is 18 or older, and then checks if they are a US citizen. Both conditions are True, so the nested if block is executed.
#Example of Complex if-elif-else with multiple conditions:
#You can also use multiple conditions with logical operators like and, or, and not.
#Example:
age = 20
student = True
if age < 18:
    print("You are a minor.")
elif age >= 18 and student:
    print("You are an adult and a student.")
else:
    print("You are an adult.")
'''Here, the elif condition checks two conditions at once: age >= 18 and student being True. Since both conditions are satisfied, the corresponding block is executed.'''
```

```python
#Describe the different types of loops in python and their use cases with examples.
'''Python provides two types of loops for repeating a block of code multiple times:
1.	for loop: Iterates over a sequence (e.g., list, tuple, string) or a range of numbers.
2.	while loop: Repeats a block of code as long as a condition is True.
Additionally, there are loop control statements like break, continue, and else used within loops to control the flow.'''
#1. for Loop
#The for loop iterates over each item in a sequence (such as a list, tuple, or string) or over a range of numbers.
#Syntax:
#for variable in sequence:
    # Block of code
#Example: Iterating Over a List
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
#Example: Using range() with a for loop
for i in range(5):
    print(i)
'''
Use Cases for for Loops:
•	Iterating over a list, tuple, string, dictionary, or set.
•	Repeating actions for a fixed number of times (e.g., iterating over a range).
•	Processing items in a collection.
'''
#2. while Loop
#The while loop repeatedly executes a block of code as long as the condition remains True. Once the condition becomes False, the loop terminates.
#Syntax:
#while condition:
    # Block of code
#Example: Printing Numbers Less Than 5
i = 0
while i < 5:
    print(i)
    i += 1
'''
Use Cases for while Loops:
•	Running a loop until a certain condition is met (e.g., waiting for user input or data processing).
•	Infinite loops that stop based on an internal event or user action (often used in real-time systems).
'''
#Example: Infinite while loop with break

while True:  # Infinite loop that will break when the user types 'quit'
    user_input = input("Type 'quit' to stop: ")
    if user_input == "quit":
        break

#In this example, the loop runs indefinitely until the user types 'quit', at which point the break statement terminates the loop.
#3. Loop Control Statements
'''
•	break: Terminates the loop completely.
•	continue: Skips the current iteration and moves to the next one.
•	else: Runs after the loop finishes, unless the loop is terminated with break.
'''
#break Example:
for i in range(10):
    if i == 5:
        break
    print(i)
#continue Example:
for i in range(5):
    if i == 2:
        continue  # Skip when i is 2
    print(i)
#else in Loops:
#The else block in a loop is executed only when the loop terminates normally (i.e., not interrupted by break).
#Example with else:
for i in range(5):
    print(i)
else:
    print("Loop finished successfully.")
#4. Nested Loops
#You can use a loop inside another loop, which is known as a nested loop.
#Example: Nested for Loop
for i in range(3):
    for j in range(2):
        print(f"i = {i}, j = {j}")
```
