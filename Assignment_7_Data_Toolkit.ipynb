1. **Demonstrate three different methods for creating identical 2D arrays in NumPy. Provide the code for each method and the final output after each method.**

3. **Explain the following terms**
*   **The difference in np.array, np.asarray and np.asanyarray.**
*   **The difference between Deep copy and shallow copy**




```python
#Difference between np.array(), np.asarray(), and np.asanyarray()
'''
np.array():

i.  Always creates a new array.
ii. It copies the data unless explicitly told not to (copy=False).
iii.If the input is already an array, it can be copied again, depending on the parameters passed.

'''
import numpy as np

# Using np.array
arr1 = np.array([1, 2, 3])
print(arr1)


'''
np.asanyarray():

i.  Similar to np.asarray(), but it preserves the subclass of the input array (if it is a subclass of ndarray).
ii. It won't convert subclassed arrays (like matrix or masked arrays) into base ndarray type.
'''

# Using np.asarray
list_data = [1, 2, 3]
arr2 = np.asarray(list_data)
print(arr2)  # Output: [1 2 3]

# Using it on an existing array
arr3 = np.asarray(arr1)
print(arr3 is arr1)


#Difference between Deep Copy and Shallow Copy
'''
Shallow Copy:

i.  A shallow copy creates a new object, but the elements inside the object are references to the same objects in the original. It does not create copies of the nested objects.
ii. Changes to mutable elements within the nested objects will affect both the original and the copied object.
In Python, shallow copies can be made using the copy() method or the copy module.
'''
import copy

original_list = [[1, 2, 3], [4, 5, 6]]
shallow_copied_list = copy.copy(original_list)

shallow_copied_list[0][0] = 99  # Modify a mutable object in the shallow copy

print("Original List:", original_list)
print("Shallow Copied List:", shallow_copied_list)


'''
Deep Copy:

i.  A deep copy creates a new object and recursively copies all objects inside, including nested objects.
ii. Changes to the copied object do not affect the original object and vice versa.
In Python, deep copies can be made using the deepcopy() method from the copy module.
'''
original_list = [[1, 2, 3], [4, 5, 6]]
deep_copied_list = copy.deepcopy(original_list)

deep_copied_list[0][0] = 99  # Modify a mutable object in the deep copy

print("Original List:", original_list)
print("Deep Copied List:", deep_copied_list)
```

14. **Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:**

  b) **Label the x-axis as 'Date'.**

  C) **Set the title of the plot as 'Temperature and Humidity Over Time'**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Create random temperature and humidity data
temperature = np.random.uniform(low=15, high=35, size=len(dates))  # Temperature between 15 and 35 degrees
humidity = np.random.uniform(low=30, high=90, size=len(dates))     # Humidity between 30% and 90%

# Create a DataFrame
data = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# a) Plot Temperature on the left y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(data['Date'], data['Temperature'], color='tab:blue', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(data['Date'], data['Humidity'], color='tab:orange', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# b) Label the x-axis as 'Date'
plt.title('Temperature and Humidity Over Time')

# c) Show the plot
plt.show()
```

15. **Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:**

  c) **Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std_dev=1

# a) Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='violet', label='Histogram')

# b) Overlay a line plot representing the normal distribution's PDF
# Calculate the PDF
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, loc=0, scale=1)  # mean=0, std_dev=1
plt.plot(x, pdf, 'green', label='Normal PDF')

# c) Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

17. **Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create two random arrays with values that can fall into all four quadrants
x = np.random.uniform(-10, 10, 100)  # Random values between -10 and 10
y = np.random.uniform(-10, 10, 100)  # Random values between -10 and 10

# Create a DataFrame to store the data
data = pd.DataFrame({'X': x, 'Y': y})

# Define a function to categorize points into quadrants
def get_quadrant(row):
    if row['X'] >= 0 and row['Y'] >= 0:
        return 'Quadrant 1'
    elif row['X'] < 0 and row['Y'] >= 0:
        return 'Quadrant 2'
    elif row['X'] < 0 and row['Y'] < 0:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

# Apply the function to create a new column for quadrants
data['Quadrant'] = data.apply(get_quadrant, axis=1)

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', palette='Set1', s=100)

# Draw lines for the axes to clearly show quadrants
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')

# Label the axes and set the title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Add a legend
plt.legend(title='Quadrant')

# Show the plot
plt.grid()
plt.show()
```

18. **With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.**

```python
from math import pi, sin
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare data for the sine wave
x = np.linspace(0, 4 * pi, 100)  # 0 to 4*pi
y = np.sin(x)                    # Sine values

# Create a Bokeh figure
p = figure(title="Sine Wave Function",
           x_axis_label='X-axis',
           y_axis_label='Y-axis',
           width=800,
           height=400)

# Add the sine wave line to the figure
p.line(x, y, line_width=2, color='blue', legend_label='y = sin(x)')

# Add grid lines
p.grid.grid_line_alpha = 0.3  # Set grid line transparency

# Show the legend
p.legend.location = "top_left"

# Output the plot to the notebook
output_notebook()
show(p)
```

19. **Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.**

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool, LinearColorMapper
from bokeh.transform import linear_cmap

# Generate random categorical data
categories = [f'Category {i}' for i in range(1, 11)]  # 10 categories
values = np.random.randint(1, 100, size=10)  # Random values between 1 and 100

# Create a ColumnDataSource for the bar chart
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a color mapper for continuous values
mapper = linear_cmap(field_name='values', palette='Viridis256', low=min(values), high=max(values))

# Create a Bokeh figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values',
           toolbar_location=None, height=400)

# Add bars to the figure with color mapping
bars = p.vbar(x='categories', top='values', width=0.9, source=source,
              line_color='white', fill_color=mapper)

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Configure visual aspects
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None

# Output the plot to the notebook
output_notebook()
show(p)
```
